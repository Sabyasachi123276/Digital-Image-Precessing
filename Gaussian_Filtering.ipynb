```python
import cv2
import numpy
from matplotlib import pyplot as plt

# Reading the image
img0 = cv2.imread('img1.bmp',)

# Blurring image by sending the ImageFilter.
# GaussianBlur predefined kernel argument
dst = cv2.GaussianBlur(img0,(5,5),cv2.BORDER_DEFAULT)

plt.subplot(121),plt.imshow(img0,cmap = 'gray')
plt.title('Original Image'), plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(dst,cmap = 'gray')
plt.title('Gaussian Filtered Image'), plt.xticks([]), plt.yticks([])

# Displaying the image
plt.show()
```
