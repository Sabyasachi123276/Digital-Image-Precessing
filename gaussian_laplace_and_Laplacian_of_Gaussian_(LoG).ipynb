The first approach to use scipy module to check the performances of multidimensional Laplace filter using Gaussian second derivatives on images.

```python
from scipy import ndimage
import cv2 as cv
import matplotlib.pyplot as plt
```

```python
img = cv.imread('img1.bmp')
```

```python
fig = plt.figure()
plt.gray()  # show the filtered result in grayscale
ax1 = fig.add_subplot(121)  # left side
ax2 = fig.add_subplot(122)  # right side
result = ndimage.gaussian_laplace(img, sigma=1)
ax1.imshow(result)
result = ndimage.gaussian_laplace(img, sigma=3)
ax2.imshow(result)
plt.show()
```

LoG on the image can be performed in the following ways:


(i) First, apply Gaussian and then Laplacian or Convolve the image with LoG kernel directly.

(ii) Find the zero crossings in the image.

(iii) Threshold the zero crossings to extract only the strong edges.

```python
# Apply Gaussian Blur
blur = cv.GaussianBlur(img,(3,3),0)

# Apply Laplacian operator in some higher datatype
laplacian = cv.Laplacian(blur,cv.CV_64F)
```

Since zero crossings is a change from negative to positive and vice-versa, so an approximate way is to clip the negative values to find the zero crossings.

```python
laplacian1 = laplacian/laplacian.max()
plt.imshow(laplacian1,cmap = 'gray')
plt.show()
```
